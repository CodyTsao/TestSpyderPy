```python
import numpy as np
city=["A", "B", "C", "D", "E"]
pos = np.arange(len(city))
barwidth=0.4

happy = [120, 180, 80, 140, -20]
happy2=[70,80,100,150,70]
plt.xticks(pos, city)
plt.bar(pos-barwidth/2,  happy, barwidth, color='blue', edgecolor='black')
plt.bar(pos+barwidth/2,  happy2, barwidth, color='green', edgecolor='blue')
plt.xlabel('City')
plt.ylabel('score')
plt.title('test')
plt.legend(['H1','H2'], loc=1)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
from matplotlib.font_manager import FontProperties
font=FontProperties(fname=r'NotoSansCJKtc-hinted\NotoSansCJKtc-Regular.otf', size=10)
mpl.rc('font', family='Noto Sans CJK TC')

#values = [8,15,30,45,50,52,55,56,57,60,70,80,90,110,120,135]
values=[98,40,62,74,58,96,25,46,38,74,68,75,34,47,55,68,63,74,75,82,8]
plt.hist(values,bins=10,range=(0,100),histtype='bar',align='mid',color='c',label='Test histograms',edgecolor='black')
plt.legend(loc=2)
plt.xlabel('成績', fontsize=16, fontproperties=font)
plt.ylabel('人數', fontsize=16, fontproperties=font)
plt.title("測試直方圖", fontproperties=font)
plt.show()
```

```python
practtime1=[1.0,3.0,2.5,1.2,7.0,6.0,5.4,4.8,6.5,7.0,7.2,6.2,8.0,4.0,5.3,5.5,3.2,1.8,0.7]
racescore1=[30,70,65,50,35,90,85,87,60,45,70,88,40,58,75,92,64,60,32]
practtime2=[1.2,2.5,4.2,4.5,4.5,4.7,3.7,3.8,3.5,2.8,2.6,2.4,7.2,6.8,6.6,6.5,6.7,7.5,7.6]
racescore2=[40,64,72,66,75,85,68,78,70,42,48,44,60,32,40,60,95,50,42]

practtime=practtime1+practtime2
racescore=racescore1+racescore2
colorarray=['g'] *19+['b']*19
#markerarray=['*']*19+['o']*19
plt.scatter(practtime, racescore, marker='*', c=colorarray)
plt.xlabel('訓練時間', fontsize=16, fontproperties=font)
plt.ylabel('競賽成績', fontsize=16, fontproperties=font)
plt.show()
```
